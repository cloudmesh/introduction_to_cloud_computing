# `nova` command in Cloudmesh

Cloudmesh provides a wrapper command for nova client tool with a switchable novarc command. You can run `nova` command to select the cloud that you want to use.

.. note:
   ! is used to run cm command in ipython notebook.

## Set cloud

You choose the cloud to run `nova` command e.g. india or icehouse on futuresystems.

```python
!cm nova set india
```

Now you are loaded **novarc** for India OpenStack.

## Novarc Contents

`info` command displays the OpenStack environment variables for the selected novarc.

```python
!cm nova info
```

```python
!cm "nova help"
```
